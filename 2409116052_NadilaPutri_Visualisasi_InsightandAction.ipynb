### Outliers

Outliers adalah nilai yang jauh berbeda dari nilai lainnya dalam dataset. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga.

Melakukan pengecekan outliers:

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dikarenakan presentase yang tidak terlalu tinggi namun merupakan variable yang penting, maka dilakukan imputasi dengan metode yang paling sesuai, yaitu Winsorizing karena tetap mempertahankan struktur data tanpa membuang nilai ekstrem.

Melakukan imputasi:

```python
columns_to_impute = ["Rating (Out of 5)", "Average Battery Life (in days)"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Melakukan pengecekan ulang:

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Distribusi Harga Fitness Tracker

```python
# Histogram (Seaborn)
plt.figure(figsize=(8, 5))
sns.histplot(df['Selling Price'], bins=20, kde=True, color='blue')
plt.title('Distribusi Harga Fitness Tracker')
plt.xlabel('Harga (INR)')
plt.ylabel('Frekuensi')
plt.show()
```

## Rata-rata Harga per Merk

```python
# Bar Chart (Seaborn)
brand_avg_price = df.groupby('Brand Name')['Selling Price'].mean().sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=brand_avg_price.index, y=brand_avg_price.values, palette="viridis")
plt.xticks(rotation=45)  # Memiringkan teks agar terbaca
plt.xlabel("Brand Name")
plt.ylabel("Average Selling Price")
plt.title("Rata-rata Harga per Merek")
plt.show()
```

## Analisis Diskon dan Dampaknya terhadap Harga

```python
# Bar Chart (Seaborn)
df_groupby = df.groupby('Brand Name')[['Original Price', 'Selling Price']].mean().sort_values(by='Original Price', ascending=False).reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(x='Brand Name', y='Original Price', data=df_groupby, color='lightgray', label='Original Price')
sns.barplot(x='Brand Name', y='Selling Price', data=df_groupby, color='blue', alpha=0.7, label='Selling Price')

plt.title('Perbandingan Harga Awal vs Harga Diskon per Merek', fontsize=14)
plt.xlabel('Merek', fontsize=12)
plt.ylabel('Harga (INR)', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Popularitas Merk Berdasarkan Jumlah Model dan Harga Rata-rata

```python
# Bar Chart + Line Chart (Seaborn)
brand_counts = df['Brand Name'].value_counts().reset_index()
brand_counts.columns = ['Brand Name', 'Model Count']

brand_avg_price = df.groupby('Brand Name')[['Selling Price']].mean().reset_index()
brand_avg_price.columns = ['Brand Name', 'Average Price']

brand_summary = brand_counts.merge(brand_avg_price, on='Brand Name')
brand_summary = brand_summary.sort_values(by='Model Count', ascending=False)

top_n = 15
brand_summary = brand_summary.head(top_n)

fig, ax1 = plt.subplots(figsize=(14, 7))

sns.barplot(x='Brand Name', y='Model Count', data=brand_summary, palette='pastel', ax=ax1)
ax1.set_ylabel('Jumlah Model', fontsize=12)
ax1.set_xlabel('Merk', fontsize=12)
ax1.set_title('Popularitas Merk: Jumlah Model vs Harga Rata-rata', fontsize=14)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
sns.lineplot(x='Brand Name', y='Average Price', data=brand_summary, color='red', marker='o', linestyle='dashed', ax=ax2)
ax2.set_ylabel('Harga Rata-rata (INR)', fontsize=12, color='red')
ax2.tick_params(axis='y', colors='red')

ax1.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
